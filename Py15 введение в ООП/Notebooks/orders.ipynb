```python
import pandas as pd

orders= pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')

orders['Product_ID']=orders['ID товара']
orders = orders.drop(['ID товара'], axis=1)

#orders.head(5)
#products.info()
merged = orders.merge(
    products,
    on='Product_ID',
    how='inner'
)

mask = merged['Оплачен'] == 'Да'
group =merged[mask].groupby('ID Покупателя')
display(group.head(10))
#display(merged.head(18))

```
