```python
import pandas as pd
melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
melb=melb_data.copy()
melb= melb.drop(['index', 'Coordinates'], axis=1)

total_rooms = melb['Rooms'] + melb['Bedroom'] + melb['Bathroom']
melb['MeanRoomsSquare'] = melb['BuildingArea'] / total_rooms

diff_area = melb['BuildingArea'] - melb['Landsize']
sum_area = melb['BuildingArea'] + melb['Landsize']
melb['AreaRatio'] = diff_area/sum_area

melb['Date'] = pd.to_datetime(melb['Date'], dayfirst=True)
melb['MonthSale'] = melb['Date'].dt.month

melb['AgeBuilding'] = melb['Date'].dt.year - melb['YearBuilt']
melb = melb.drop('YearBuilt', axis=1)

melb['WeekdaySale']=melb['Date'].dt.dayofweek

mask=(melb['WeekdaySale']==6) | (melb['WeekdaySale']==5)
#print(melb[mask].shape[0])

# На вход данной функции поступает строка с адресом.
def get_street_type(address):
# Создаём список географических пометок exclude_list.
    exclude_list = ['N', 'S', 'W', 'E']
# Метод split() разбивает строку на слова по пробелу.
# В результате получаем список слов в строке и заносим его в переменную address_list.
    address_list = address.split(' ')
# Обрезаем список, оставляя в нём только последний элемент,
# потенциальный подтип улицы, и заносим в переменную street_type.
    street_type = address_list[-1]
# Делаем проверку на то, что полученный подтип является географической пометкой.
# Для этого проверяем его на наличие в списке exclude_list.
    if street_type in exclude_list:
# Если переменная street_type является географической пометкой,
# переопределяем её на второй элемент с конца списка address_list.
        street_type = address_list[-2]
# Возвращаем переменную street_type, в которой хранится подтип улицы.
    return street_type
street_types = melb['Address'].apply(get_street_type)
#display(street_types)
#посмотрим на частоту каждого подтипа улицы с помощью метода value_counts:

#display(street_types.value_counts())

#мы хотим отобрать десять наиболее популярных подтипов. 
# Извлечём их названия с помощью атрибута index, 
# а результат занесём в переменную popular_stypes:
popular_stypes =street_types.value_counts().nlargest(10).index
melb['StreetType'] = street_types.apply(lambda x: x if x in popular_stypes else 'other')

#print(melb[mask]['Price'].mean())
melb['SellerG'].value_counts()
sellers=melb['SellerG'].value_counts().nlargest(49).index
#display(sellers)
melb['SellerG']= melb['SellerG'].apply(lambda x: x if x in sellers else 'other')
mask_n = melb['SellerG']=='Nelson'
mask_o = melb['SellerG']=='other'
n_min= melb[mask_n]['Price'].min()
o_min= melb[mask_o]['Price'].min()
#print(n_min/o_min)

popular_suburb=melb['Suburb'].value_counts().nlargest(119)
melb['Suburb']= melb['Suburb'].apply(lambda x: x if x in popular_suburb else 'other')
melb['Suburb']= melb['Suburb'].astype('category')

melb.info()
```
